```python
import pandas as pd
```

```python
data= pd.read_csv('main.csv' , index_col= 'Year')
data.head(2)
```

```python
data=data.drop(columns='Total')  #output needs no total
data.head(2)
```

```python
data1= data.groupby((data.index//10)*10).sum()
data1
```

```python
data1['Population'] = data['Population'].groupby((data.index//10)*10).max()   #because the last year of decade will be the population of that decade 
data1
```

```python
data1.to_csv('output1.csv')
```
